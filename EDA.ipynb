```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('HR_Cleaned.csv')

# Attrition distribution
sns.countplot(x='Attrition', data=df)
plt.title("Attrition Distribution")

# Attrition by Department
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title("Attrition by Department")

plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation (Numeric Only)")
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set3')
plt.title("Attrition vs Monthly Income")
plt.xlabel("Attrition (0 = No, 1 = Yes)")
plt.ylabel("Monthly Income")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', kde=True, multiple='stack', palette='Set2')
plt.title("Attrition vs Years at Company")
plt.xlabel("Years at Company")
plt.ylabel("Number of Employees")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age', hue='Attrition', bins=20, multiple='stack', palette='coolwarm')
plt.title("Attrition by Age Group")
plt.xlabel("Age")
plt.ylabel("Employee Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df, palette='husl')
plt.title("Attrition vs Job Satisfaction")
plt.xlabel("Job Satisfaction (1=Low, 4=High)")
plt.ylabel("Employee Count")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='Set1')
plt.title("Attrition vs Overtime")
plt.xlabel("OverTime (0 = No, 1 = Yes)")
plt.ylabel("Employee Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='JobRole', hue='Attrition', data=df, palette='Set3')
plt.title("Attrition by Job Role")
plt.xlabel("Employee Count")
plt.ylabel("Job Role")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='EnvironmentSatisfaction', hue='Attrition', data=df, palette='cubehelix')
plt.title("Attrition vs Environment Satisfaction")
plt.xlabel("Environment Satisfaction (1=Low, 4=High)")
plt.ylabel("Employee Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='BusinessTravel', hue='Attrition', data=df, palette='pastel')
plt.title("Attrition by Business Travel Frequency")
plt.xlabel("Business Travel Category")
plt.ylabel("Employee Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df, palette='viridis')
plt.title("Attrition vs Work-Life Balance")
plt.xlabel("Work-Life Balance (1=Poor, 4=Excellent)")
plt.ylabel("Employee Count")
plt.show()
```

```python
sns.pairplot(df[['Age', 'MonthlyIncome', 'YearsAtCompany', 'Attrition']], hue='Attrition', diag_kind='kde', palette='husl')
plt.suptitle("Pairwise Relationships in HR Data", y=1.02)
plt.show()
```
