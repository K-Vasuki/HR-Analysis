```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Optional for nice styling
sns.set(style="whitegrid")
```

```python
df = pd.read_excel('Documents/DA_projects/HR-Employee.xlsx')
df.head()
```

```python
df.info()

```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Age Group
df['AgeGroup'] = pd.cut(df['Age'], bins=[18, 29, 40, 60], labels=['Below 30', '30-40', 'Above 40'])

# Salary Band
df['SalaryBand'] = pd.cut(df['MonthlyIncome'],
                          bins=[0, 3000, 8000, df['MonthlyIncome'].max()],
                          labels=['Low', 'Medium', 'High'])
```

```python
df.head()
```

```python
# Total employees
total_employees = df.shape[0]

# Attrition count
attrition_count = df[df['Attrition'] == 'Yes'].shape[0]

# Attrition rate
attrition_rate = round(attrition_count / total_employees * 100, 2)

# Avg age of attrited employees
avg_age_attrited = df[df['Attrition'] == 'Yes']['Age'].mean()

print(f"Total Employees: {total_employees}")
print(f"Attrition Count: {attrition_count}")
print(f"Attrition Rate: {attrition_rate}%")
print(f"Avg Age of Attrited Employees: {avg_age_attrited:.2f}")
```

```python
sns.countplot(data=df, x='Attrition')
plt.title("Attrition Count")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Department', hue='Attrition', palette='Pastel1')
plt.title("Attrition by Department")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='AgeGroup', hue='Attrition', palette='Set3')
plt.title("Attrition by Age Group")
plt.show()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.title("Income Distribution by Attrition")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='OverTime', hue='Attrition', palette='Set2')
plt.title("Attrition by OverTime Status")
plt.show()
```

```python
plt.figure(figsize=(14, 10))
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```
